# Bonus: Challenge 2: creating functions on a separate `py` file

Put all the data cleaning and formatting steps into functions, and create a main function that performs all the cleaning and formatting.

Write these functions in separate .py file(s). By putting these steps into functions, we can make the code more modular and easier to maintain.

*Hint: autoreload module is a utility module in Python that allows you to automatically reload modules in the current session when changes are made to the source code. This can be useful in situations where you are actively developing code and want to see the effects of changes you make without having to constantly restart the Python interpreter or Jupyter Notebook kernel.*

```python
import numpy as np
import pandas as pd

def load_data(url):
    data = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
    data.reset_index(drop=True, inplace=True)
    data.columns = [c.lower().strip() for c in data.columns]
    return data

def rename_columns(data, columns_map):
    data.rename(columns=columns_map, inplace=True)

def replace_values(data, column, replacements):
    data[column].replace(replacements, inplace=True)

def clean_customer_lifetime_value(data, column):
    data[column] = data[column].str.replace('%', '').astype(float)

def convert_numeric_columns(data):
    float_cols = data.select_dtypes(include=['float64']).columns
    data[float_cols] = data[float_cols].astype(int)

def main():
    # Load and prepare data
    url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
    insurance_data = load_data(url)

    # Rename columns
    rename_columns(insurance_data, {'st': 'state'})

    # Replace values
    state_replacements = {'WA': 'Washington', 'AZ': 'Arizona', 'Cali': 'California'}
    replace_values(insurance_data, 'state', state_replacements)

    gender_replacements = {'Femal': 'F', 'female': 'F', 'Male': 'M'}
    replace_values(insurance_data, 'gender', gender_replacements)

    education_replacements = {'Bachelors': 'Bachelor'}
    replace_values(insurance_data, 'education', education_replacements)

    clean_customer_lifetime_value(insurance_data, 'customer lifetime value')

    vehicle_class_replacements = {'Sports Car': 'Luxury', 'Luxury Car': 'Luxury'}
    replace_values(insurance_data, 'vehicle class', vehicle_class_replacements)

    complaints_replacements = {
        '1/0/00': '0', '1/1/00': '1', '1/2/00': '2', 
        '1/3/00': '3', '1/4/00': '4', '1/5/00': '5'
    }
    replace_values(insurance_data, 'number of open complaints', complaints_replacements)

    # Drop NA and duplicates
    insurance_data.dropna(inplace=True)
    insurance_data.drop_duplicates(inplace=True)
    insurance_data.reset_index(drop=True, inplace=True)

    # Convert numeric columns to int
    convert_numeric_columns(insurance_data)

    # Output results
    print(insurance_data.shape)
    print(insurance_data.dtypes)
    print(insurance_data.duplicated())
    print(insurance_data[insurance_data.duplicated()])
    print(insurance_data.head())

if __name__ == "__main__":
    main()
```
